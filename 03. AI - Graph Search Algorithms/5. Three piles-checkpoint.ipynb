```python
from easyAI import TwoPlayersGame, Human_Player, AI_Player, Negamax

class ThreePiles( TwoPlayersGame ):
    """ In turn, the players remove one, two or three bones from a
    pile of bones. The player who removes the last bone loses. """

    def __init__(self, players):
        self.players = players
        self.piles = [5, 5, 5] # start with 20 bones in the pile1
        self.nplayer = 1 # player 1 starts

    def possible_moves(self):
        result = []
        for i in range(3):
            j = 1
            while (j <= self.piles[i]) and (j <= 3):
                result.append(str(i) + ',' + str(j))
                j += 1
        # print(result)
        return result

    def make_move(self,move):
        self.piles[int(move.split(',')[0])] -= int(move.split(',')[1]) # remove bones.
        
    def win(self): 
        return (self.piles[0] <= 0) and (self.piles[1] <= 0) and (self.piles[2] <= 0) # opponent took the last bone ?

    def is_over(self): 
        return self.win() # Game stops when someone wins.

    def show(self):
        print("%d bones left on pile 0" % self.piles[0])
        print("%d bones left on pile 1" % self.piles[1])
        print("%d bones left on pile 2" % self.piles[2])
        
    def scoring(self): 
        return 100 if self.win() else 0 # For the AI

# Start a match (and store the history of moves when it ends)
ai = Negamax(6) # The AI will think 6 moves in advance
game = ThreePiles( [ Human_Player(), AI_Player(ai) ] )
history = game.play()
```
